```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt



# Convertir la columna de fechas a tipo datetime tipo de dato para poder procesar la informacion
df['fecha_reporte_visible'] = pd.to_datetime(df['fecha_reporte_visible'])



# Agregar columnas de mes y año al DataFrame
df['Mes'] = df['fecha_reporte_visible'].dt.month #extrae el componente del mes de cada fecha en la columna 'fecha_reporte_visible' y asigna esos valores a una nueva columna llamada 'Mes' en el dataFrame df.
df['Año'] = df['fecha_reporte_visible'].dt.year #extrae el componente del año de cada fecha en 'fecha_reporte_visible' y asigna esos valores a una nueva columna llamada 'Año' en el dataFrame df.



# Crear un DataFrame con los datos de conteo por mes
df_conteos_mes = df.groupby(['Año', 'Mes']).size().reset_index(name='Denuncias') #groupby para agrupar el DataFrame df por las columnas 'Año' y 'Mes'
#y luego utiliza la función size() para contar la cantidad de registros en cada grupo por ultimo regresa el nombre a "denuncias"



# Calcular la diferencia entre los conteos de los meses
df_conteos_mes['Diferencia'] = df_conteos_mes['Denuncias'].diff() #Esta línea de código calcula la diferencia entre los valores sucesivos en la columna 'Denuncias' del
#DataFrame df_conteos_mes y agrega esos resultados como una nueva columna llamada 'Diferencia'.



# Establecer la variable objetivo como la tendencia de las denuncias (aumento o disminución)
#"Tendencia' está siendo utilizada para representar si hay un aumento (1) o no (0) en el número de denuncias entre dos meses consecutivos.
df_conteos_mes['Tendencia'] = df_conteos_mes['Diferencia'].apply(lambda x: 1 if x > 0 else 0)



# Separar los datos en características (X) y etiquetas (y)
x = df_conteos_mes[['Año', 'Mes']].values # contiene año y mes
y = df_conteos_mes['Tendencia'].values #contiene datos binarios



# Dividir los datos en conjunto de entrenamiento y prueba


x_entrenamiento, x_prueba, y_entrenamiento, y_prueba = train_test_split(x, y, test_size=0.2, random_state=24)
#train_test_split: Esta función de scikit-learn se utiliza para dividir los datos en conjuntos de entrenamiento y prueba de manera aleatoria. En este caso,
#el 80% de los datos se asignan al conjunto de entrenamiento (X_entrenamiento, y_entrenamiento),
# y el 20% restante se asigna al conjunto de prueba (X_prueba, y_prueba).
#random_state: La importancia radica en que si utilizas la misma semilla, obtendrás la misma secuencia de números pseudoaleatorios
#cada vez que ejecutes el código. Esto es esencial para la reproducibilidad de los resultados.




# Normalizar los datos


scaler = MinMaxScaler()#Aquí, se crea una instancia de la clase MinMaxScaler. Este objeto se utilizará para ajustar y transformar los datos.


x_entrenamiento = scaler.fit_transform(x_entrenamiento) #La función fit_transform realiza dos operaciones. Primero, ajusta el escalador (scaler)
#a los datos de entrenamiento (x_entrenamiento), calculando los valores mínimos y máximos necesarios
#para la normalización. Luego, transforma los datos de entrenamiento aplicando la normalización.
#Después de este paso, x_entrenamiento contiene los datos normalizados.


x_prueba = scaler.transform(x_prueba) #La función transform aplica la misma normalización que se ajustó al conjunto de entrenamiento
#a los datos de prueba (x_prueba). Esto es importante para mantener la coherencia en la escala entre los datos de entrenamiento y prueba.



# Generar un tercio de datos de ruido


ruido = np.random.rand(len(x) // 2, 2) #Se utiliza la función np.random.rand para generar un conjunto de datos aleatorios de forma
#(len(x) // 2, 2). Esto significa que estás generando datos de ruido con la mitad del tamaño de tu conjunto de datos original (x).
#Cada fila de ruido contendrá dos valores aleatorios entre 0 y 1.


x_ruido = scaler.transform(ruido)#Se aplica la transformación de escala (transform) utilizando el mismo escalador (scaler)
#que se ajustó a los datos de entrenamiento originales.
#La transformación se aplica al conjunto de datos de ruido generado, y x_ruido contendrá estos datos de ruido normalizados.

# Combinar datos originales y de ruido


x_entrenamiento_con_ruido = np.vstack((x_entrenamiento, x_ruido)) #np.vstack se utiliza para apilar verticalmente
#los conjuntos de datos x_entrenamiento (datos de entrenamiento originales) y x_ruido (datos de ruido).
#Después de esta línea de código, x_entrenamiento_con_ruido contendrá tanto las características originales como el ruido.


etiquetas_ruido = np.random.randint(2, size=len(x_ruido))  # Genera etiquetas aleatorias binarias (0 o 1)
#np.random.randint(2, size=len(x_ruido)) genera una secuencia de números enteros aleatorios en el rango [0, 2), es decir, puede generar 0 o 1.
#La longitud de esta secuencia es la misma que la cantidad de datos de ruido en x_ruido.


y_entrenamiento_con_ruido = np.hstack((y_entrenamiento, etiquetas_ruido))#np.hstack, estás apilando horizontalmente las etiquetas originales de
#entrenamiento (y_entrenamiento) con las nuevas etiquetas de ruido generadas (etiquetas_ruido). Esto significa que y_entrenamiento_con_ruido
#contendrá las etiquetas originales y, al final, las etiquetas de ruido generadas aleatoriamente.




# Ajuste de la arquitectura del modelo para regresión logística


modelo_regresion_logistica = Sequential() #Se crea un objeto de modelo secuencial (Sequential).
#Este tipo de modelo se utiliza para apilar capas en orden secuencial, una encima de la otra.


modelo_regresion_logistica.add(Dense(1, activation='sigmoid', input_dim=2)) #1: Indica que la capa tendrá una sola neurona.
#En problemas de regresión logística binaria, una única neurona es común en la capa de salida.
#activation='sigmoid': Especifica que se utilizará la función de activación sigmoide en la salida.
#La función sigmoide convierte la salida en un rango entre 0 y 1, lo que es adecuado para problemas de clasificación binaria.
#input_dim=2: Indica que la capa de entrada tiene dos dimensiones. En este caso, se espera que las características de
#entrada tengan dos dimensiones.



# Ajuste de la tasa de aprendizaje y función de pérdida


modelo_regresion_logistica.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
#learning_rate=0.001: Establece la tasa de aprendizaje del optimizador,
#que controla el tamaño de los pasos de actualización de los pesos durante el entrenamiento.
#Se utiliza la función de pérdida binary_crossentropy, que es comúnmente utilizada para problemas de clasificación binaria.
#Esta función de pérdida es apropiada cuando el modelo produce salidas en forma de probabilidades y se espera que realice una clasificación binaria.
#Se especifica que la métrica a seguir durante el entrenamiento es la precisión (accuracy),
#que es una medida comúnmente utilizada para evaluar el rendimiento de modelos de clasificación.



# Ajuste del tamaño del lote y el número de épocas


historial_entrenamiento_regresion = modelo_regresion_logistica.fit(x_entrenamiento_con_ruido,
                                                                   y_entrenamiento_con_ruido,
                                                                   epochs=10000, batch_size=4,
                                                                   validation_data=(x_prueba, y_prueba))
#x_entrenamiento_con_ruido: Contiene las características de entrenamiento ampliadas con datos de ruido.
#y_entrenamiento_con_ruido: Contiene las etiquetas de entrenamiento ampliadas con ceros y unos para los datos de ruido.
#epochs indica la cantidad de veces que se recorrerán todos los datos de entrenamiento durante el proceso de entrenamiento.
#validation_data especifica el número de muestras que se utilizarán en cada iteración del proceso de entrenamiento.



# Obtener el último año y mes en los datos


ultimo_año = df_conteos_mes['Año'].max()#df_conteos_mes['Año'].max(): Utiliza el método max()
#para encontrar el valor máximo en la columna 'Año' del DataFrame df_conteos_mes.
#ultimo_año: Almacena el año máximo, que representa el último año presente en los datos.


ultimo_mes = df_conteos_mes[df_conteos_mes['Año'] == ultimo_año]['Mes'].max() #df_conteos_mes[df_conteos_mes['Año'] == ultimo_año]:
#Filtra las filas del DataFrame donde el año es igual al último año.
#['Mes'].max(): Utiliza el método max() en la columna 'Mes' del DataFrame resultante para encontrar el mes máximo.
#ultimo_mes: Almacena el último mes del último año presente en los datos.
#Estas variables (ultimo_año y ultimo_mes) se utilizan posteriormente para generar datos para los
#12 meses posteriores en el conjunto de prueba y realizar predicciones con el modelo entrenado.




# Generar datos para los 12 meses posteriores


meses_posteriores = np.array([[ultimo_año, ultimo_mes + i] for i in range(1, 13)])#Se utiliza una comprensión de lista
#para generar una matriz de NumPy (np.array) que contiene pares de años y meses.
#ultimo_año: Representa el año del último registro en tus datos.
#ultimo_mes: Representa el último mes del último año en tus datos.
#for i in range(1, 13): Genera valores de 1 a 12 para representar los meses posteriores.


x_posteriores = scaler.transform(meses_posteriores)#scaler.transform: Utiliza el escalador
#(scaler) previamente ajustado en los datos de entrenamiento para normalizar los nuevos datos generados (meses_posteriores).
#X_posteriores: Almacena los datos normalizados para los meses posteriores.
#Estos datos normalizados se utilizan más adelante para realizar predicciones sobre la tendencia esperada en los
#12 meses siguientes utilizando el modelo de regresión logística entrenado.




# Realizar predicciones para los 12 meses posteriores


predicciones_posteriores_regresion = modelo_regresion_logistica.predict(x_posteriores)
#modelo_regresion_logistica: Es el modelo de regresión logística que has entrenado previamente.
#predict(X_posteriores): Utiliza el método predict del modelo para realizar predicciones sobre los datos normalizados
#de los 12 meses posteriores (X_posteriores).




# Desnormalizar las predicciones para obtener el número estimado de reportes


denuncias_estimadas_posteriores_regresion = predicciones_posteriores_regresion * df_conteos_mes['Denuncias'].max()
#predicciones_posteriores_regresion: Contiene las predicciones del modelo para los 12 meses posteriores (valores normalizados entre 0 y 1).
#df_conteos_mes['Denuncias'].max(): Obtiene el valor máximo de la columna 'Denuncias' en el DataFrame original. Esto proporciona una escala
#para desnormalizar las predicciones.
#La multiplicación desnormaliza las predicciones multiplicando cada valor predicho por el máximo número de denuncias registrado en el
#DataFrame original. Esto convierte las predicciones de nuevo a la escala original, representando así el número estimado de denuncias
#para cada uno de los 12 meses siguientes.




# Imprimir el número estimado de reportes para cada mes
for i, estimado in enumerate(denuncias_estimadas_posteriores_regresion):
    print(f"Mes {i + 1}: Número estimado de reportes (Desnormalizacion de datos):{estimado[0]:.2f}")
#enumerate(denuncias_estimadas_posteriores_regresion): Enumera sobre la lista denuncias_estimadas_posteriores_regresion,
#proporcionando tanto el índice (i) como el valor (estimado).
#ultimo_mes + i + 1: Calcula el mes correspondiente para cada iteración.
#estimado[0]: Accede al valor de la predicción en la posición 0 del array estimado
#(ya que parece que denuncias_estimadas_posteriores_regresion contiene arrays unidimensionales).
#:.2f: Formatea el número como un flotante con dos decimales.





# Imprimir las probabilidades de las predicciones
for i, probabilidad in enumerate(predicciones_posteriores_regresion):
    print(f"Mes {i + 1}: Probabilidad de aumento (Regresión Logística): {probabilidad[0]:.4f}")
#enumerate(predicciones_posteriores_regresion): Enumera sobre la lista predicciones_posteriores_regresion,
#proporcionando tanto el índice (i) como el valor (probabilidad).
#probabilidad[0]: Accede al valor de la probabilidad en la posición 0 del array probabilidad
#(ya que parece que predicciones_posteriores_regresion contiene arrays unidimensionales).
#:.3f: Formatea el número como un flotante con cuatro decimales.


# Graficar los resultados
plt.figure(figsize=(10, 5))
plt.plot(df_conteos_mes['Año'] + df_conteos_mes['Mes'] / 12, df_conteos_mes['Denuncias'], 'o-', label='Datos históricos')
plt.plot(ultimo_año + (ultimo_mes + np.arange(1, 13)) / 12, denuncias_estimadas_posteriores_regresion, 's--', label='Predicciones (Regresión Logística)')
plt.title('Número Estimado de Reportes')
plt.xlabel('Año')
plt.ylabel('Número de Reportes')
plt.legend()
plt.show()
```
